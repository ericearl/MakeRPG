```python
import json
import re
import yaml

yaml_file = 'aces_and_eights/system_stats_skills.yaml'
with open(yaml_file,'r') as f:
    tree = yaml.load(f)
```

```python
print(json.dumps(tree, indent=2))
```

```python
line = tree['stats']['INTELLIGENCE']['set']
```

```python
parseable_re = re.compile(r'(stats\||skills\||points\||defaults\||options\||\d+d\d+|\(\ .+\ \))')
dice_re = re.compile(r'(\d+)d(\d+)')
number_re = re.compile(r'\ \d+\ ')
lookup_re = re.compile(r'(stats|skills|points|defaults|options)\|([\w\ ])+')
word_re = re.compile(r'\w+')
open_p_re = re.compile(r'\(\ ')
close_p_re = re.compile(r'\ \)')
comparison_re = re.compile(r'\ (<|>|<=|>=|==|!=|<>|><)\ ')
logical_func_re = re.compile(r'(min|max)\(\ ')
math_func_re = re.compile(r'(exact|round|roundup|rounddown)\(\ ')
logical_re = re.compile(r'\ \)\ (AND|OR)\ \(\ ')
math_re = re.compile(r'\ [\*\/\+\-]\ ')

dice          = re.finditer(dice_re, line)
numbers       = re.finditer(number_re, line)
lookups       = re.finditer(lookup_re, line)
words         = re.finditer(word_re, line)
open_ps       = re.finditer(open_p_re, line)
close_ps      = re.finditer(close_p_re, line)
comparisons   = re.finditer(comparison_re, line)
logical_funcs = re.finditer(logical_func_re, line)
math_funcs    = re.finditer(math_func_re, line)
logicals      = re.finditer(logical_re, line)
maths         = re.finditer(math_re, line)
```

```python
# ORDER OF OPERATIONS
# 1. option
# 2. default
# 3. point
# 4. stat
# 5. skill
# 6. parenthesis
# 7. multiplication
# 8. division
# 9. addition
# 10. subtraction

print(line)
dice_matches = [x for x in dice]
numbers_matches = [x for x in numbers]
lookups_matches = [x for x in lookups]
words_matches = [x for x in words]
open_ps_matches = [x for x in open_ps]
close_ps_matches = [x for x in close_ps]
comparisons_matches = [x for x in comparisons]
logical_funcs_matches = [x for x in logical_funcs]
math_funcs_matches = [x for x in math_funcs]
logicals_matches = [x for x in logicals]
maths_matches = [x for x in maths]

matches = dice_matches+numbers_matches+lookups_matches+words_matches+open_ps_matches+close_ps_matches+comparisons_matches+logical_funcs_matches+math_funcs_matches+logicals_matches+maths_matches
```

```python
for x in matches:
    print(x)
```
